Dropping irrelevant data 


```python
df1.drop(["company","experience","education","industry","jobdescription","jobid","jobtitle","numberofpositions","site_name","uniq_id","postdate"],inplace=True,axis=1)
df1
```

Finding the text data in payrate and converting the string into zero

```python
for i in range(0,21999):
    a=df1['payrate'][i]
    if (type(a)==str and a[0].isalpha()):
        df1['payrate'][i]=0 
df1        
```

taking the lower values of all the payrates

```python
for i in range(0,22000):
    z=""
    x=df1["payrate"][i]
    if(type(x)==str):
        for j in range(0,len(x)):
            if(x[j].isdigit()):
                z=z+x[j]
            if(x[j]=="-"):
                break
        df1["payrate"][i]=z  
df1        
```

```python
df1["payrate"].fillna(0, inplace = True)
```
