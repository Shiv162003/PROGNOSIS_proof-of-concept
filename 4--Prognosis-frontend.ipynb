```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing, svm
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import preprocessing
```

```python
profession=input("enter your desired profession ")
df=pd.read_csv(profession+".csv")
# label_encoder object knows how to understand word labels.
label_encoder = preprocessing.LabelEncoder()
  
# Encode labels in column 'species'.
df['skills/dates']= label_encoder.fit_transform(df['skills/dates'])
  
df['skills/dates'].unique()
first = list(df.columns)
first
l=[]
for i in range (1,11):
    x = np.arange(df['skills/dates'].size)
    fit = np.polyfit(x, df[first[i]], deg=1)
    print ("Slope : " + str(fit[0]))
    l.append(str(fit[0]))
    print ("Intercept : " + str(fit[1]))
    #Fit function : y = mx + c [linear regression ]
    fit_function = np.poly1d(fit)
    #Linear regression plot
    plt_1=plt.figure(figsize=(7,4))

    plt.plot(df['skills/dates'], fit_function(x))
    #Time series data plot
    plt.plot(df['skills/dates'], df[first[i]])
    plt.xticks(rotation=90);
    plt.xlabel('Dates')
    plt.ylabel('trends')
    plt.title(first[i])
    print(plt.show())
final_list2=l
final_list=first
for i in range (0,10):
    for j in range(0,9-i):
        if(final_list2[j]>final_list2[j+1]):
            temp=final_list2[j]
            final_list2[j]=final_list2[j+1]
            final_list2[j+1]=temp
            temp=final_list[j]
            final_list[j]=final_list[j+1]
            final_list[j+1]=temp
final_list.reverse()
final_list2.reverse()    
print(final_list[0:3])
print(final_list2[0:3])    
```
