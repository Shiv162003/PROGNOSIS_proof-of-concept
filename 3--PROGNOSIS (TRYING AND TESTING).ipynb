```python
#pridicting the top 3 skills according to the sum
final_list = []
final_list2=[]
for i in range(0, 3):
    max1 = 0   
    for j in range(len(l)):    
        if l[j] > max1:
            max1 = l[j];
            x=j;
    l.remove(max1);
    final_list.append(max1)
    final_list2.append(x)  
final_list2[0]= final_list2[0]+1    
final_list2[1]= final_list2[1]+2
final_list2[2]= final_list2[2]+3
for i in range(0,3):
    print(df.columns[final_list2[i]])
    print(final_list[i])
```

```python
final_list =[]
final_list2=[]
for i in range (1,11):
    print(first[i])
    final_list.append(first[i])
    df_binary = df[['skills/dates', first[i]]] 
    # Taking only the selected two attributes from the dataset
    df_binary.columns = ['skills/dates', first[i]]
    df_binary.fillna(method ='ffill', inplace = True)
    X = np.array(df_binary['skills/dates']).reshape(-1, 1)
    y = np.array(df_binary[first[i]]).reshape(-1, 1)
    # Separating the data into independent and dependent variables
    # Converting each dataframe into a numpy array 
    # since each dataframe contains only one column
    df_binary.dropna(inplace = True) 
    # Dropping any rows with Nan values
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
    # Splitting the data into training and testing data
    regr = LinearRegression()
    regr.fit(X_train, y_train)
    print(regr.score(X_test, y_test))
    final_list2.append(regr.score(X_test, y_test))
    print()
    
```

```python
print(final_list)
print(final_list2)
```

```python
# prdicting the top 3 skills according to the regression coffecient
for i in range (0,10):
    for j in range(0,9-i):
        if(final_list2[j]>final_list2[j+1]):
            temp=final_list2[j]
            final_list2[j]=final_list2[j+1]
            final_list2[j+1]=temp
            temp=final_list[j]
            final_list[j]=final_list[j+1]
            final_list[j+1]=temp
final_list.reverse()
final_list2.reverse()    
print(final_list[0:3])
print(final_list2[0:3])
```

```python

x = np.arange(df['skills/dates'].size)
fit = np.polyfit(x, df['Oracle'], deg=1)
print ("Slope : " + str(fit[0]))
print ("Intercept : " + str(fit[1]))
#Fit function : y = mx + c [linear regression ]
fit_function = np.poly1d(fit)
#Linear regression plot
plt_1=plt.figure(figsize=(20,9))

plt.plot(df['skills/dates'], fit_function(x))
#Time series data plot
plt.plot(df['skills/dates'], df['Oracle'])
plt.xticks(rotation=90);
plt.xlabel('Dates')
plt.ylabel('trends')
plt.title('Oracle')
plt.show()
```
