```python
x = np.arange(df['skills/dates'].size)
fit = np.polyfit(x, df['CSS'], deg=1)
print ("Slope : " + str(fit[0]))
print ("Intercept : " + str(fit[1]))
#Fit function : y = mx + c [linear regression ]
fit_function = np.poly1d(fit)
#Linear regression plot
plt_1=plt.figure(figsize=(20,9))

plt.plot(df['skills/dates'], fit_function(x))
#Time series data plot
plt.plot(df['skills/dates'], df['CSS'])
plt.xticks(rotation=90);
plt.xlabel('Dates')
plt.ylabel('trends')
plt.title('CSS')
plt.show()
```

```python
# libraries
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
 
# Data
fig, ax = plt.subplots(figsize=(20,10))

# multiple line plots
plt.plot( 'skills/dates', 'Azure', data=df, marker='o', markerfacecolor='blue', markersize=12, color='skyblue', linewidth=4)
plt.plot( 'skills/dates', 'Agile', data=df, marker='', color='olive', linewidth=2)
plt.plot( 'skills/dates', 'JavaScript', data=df, marker='', color='olive', linewidth=2, linestyle='dashed')
plt.xticks(rotation=90);
# show legend
plt.legend()

# show graph
plt.show()
```
